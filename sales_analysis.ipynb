```python
# üìä Sales Data Analysis with Pandas and Matplotlib

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Optional: Suppress warnings (like emoji font issues)
warnings.filterwarnings("ignore")

# Set seaborn styling
sns.set_style("whitegrid")
```

```python
# Step 2: Load the Dataset
try:
    df = pd.read_csv("sales_data.csv")
    print("‚úÖ Dataset loaded successfully!\n")
except FileNotFoundError:
    print("‚ùå Error: sales_data.csv not found.")

# Display first few rows
df.head()
```

```python
# Step 3: Explore the Dataset
print("üìå Dataset Info:\n")
df.info()

print("\nüìå Missing Values:\n")
print(df.isnull().sum())
```

```python
# Step 4: Clean the Dataset (if needed)
# Drop rows with missing values (adjust as needed)
df.dropna(inplace=True)
```

```python
# Step 5: Convert Date column to datetime & sort
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values("Date", inplace=True)
```

```python
# Step 6: Basic Statistical Summary
df.describe()
```

```python
# Step 7: Grouping - Average Sales per Region
region_avg_sales = df.groupby("Region")["Sales"].mean()
region_avg_sales
```

```python
# Line Chart - Sales Over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Sales"], marker='o', linestyle='-', color='blue')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart - Average Sales by Region
region_avg_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Sales by Region")
plt.ylabel("Average Sales")
plt.xlabel("Region")
plt.tight_layout()
plt.show()

```

```python
# Histogram - Distribution of Sales
plt.hist(df["Sales"], bins=8, color="green", edgecolor='black')
plt.title("Sales Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot - Sales vs Profit
plt.scatter(df["Sales"], df["Profit"], color="purple")
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```
